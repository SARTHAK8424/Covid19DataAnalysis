```python
#Import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset
dataset = pd.read_csv('/content/covid19_Confirmed_dataset.csv')
dataset.head()
```

```python
dataset.shape
```

```python
# delete the useless columns
df = dataset.drop(["Lat","Long"], axis = 1 , inplace = True)
```

```python
dataset.head()
```

```python
#aggregate the rows by the country
corona_dataset_aggregated = dataset.groupby("Country/Region").sum()
```

```python
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
# Strip whitespaces in column names
corona_dataset_aggregated.columns = corona_dataset_aggregated.columns.str.strip()

# Only set index if 'Country/Region' is still a column
if 'Country/Region' in corona_dataset_aggregated.columns:
    corona_dataset_aggregated.set_index("Country/Region",inplace=True)
```

```python
#visualise data related to a country

corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["India"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
#calculate a good measure
corona_dataset_aggregated.loc["China"].plot()
```

```python
corona_dataset_aggregated.loc["China"][:3].plot()
```

```python
#calculate the first derivate of the curve
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
#maximun infection rate
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["India"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for country in countries:
  max_infection_rates.append(corona_dataset_aggregated.loc[country].diff().max())
corona_dataset_aggregated["Max_infection_rates"] = max_infection_rates
```

```python
corona_dataset_aggregated.head()
```

```python
#create a new dataframe
corona_data = pd.DataFrame(corona_dataset_aggregated["Max_infection_rates"])
```
